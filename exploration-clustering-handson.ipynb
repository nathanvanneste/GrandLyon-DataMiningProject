This value can be used for finding the optimal number of clusters using the *Elbow method*. To do that let's vary the number of clusters `k`, apply k-means algorithm and store the corresponding intertia values. Then, let's plot the result and determine the best `k`.

**QUESTIONS**

* Find the optimal `k` using Elbow method

````python
# ANSWER
# Initialisation de la liste pour stocker les inerties
inertia = []
k_range = range(1, 11)  # On teste de 1  10 clusters

# Boucle pour calculer l'inertie pour chaque nombre de clusters
for k in k_range:
    # On cre le modle
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    # On entrane le modle sur les donnes standardises
    kmeans.fit(scaled_data_df)
    # On stocke l'inertie obtenue
    inertia.append(kmeans.inertia_)

# Cration du graphique (Plot)
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()
````

## Task #5: Cluster Evaluation using Silhouette Coefficient

To evaluate the quality of the clustering, we can use **Silhouette Coefficient**. The Silhouette Coefficient for a sample is given by $(b - a) / max(a, b)$ where `b` is the distance between a sample and the nearest cluster that the sample is not a part of, and `a` is the mean intra-cluster distance (i.e. the mean distance between a sample and all other samples in the same cluster). 

The silhouette score ranges from -1 to 1 and indicates how well each data point fits within its assigned cluster:

* Score near +1 means:
    - The data point is far from neighboring clusters
    - The point is well-matched to its cluster
    - Indicates very distinct, well-separated clustering
* Score near 0 means:
    - The data point is close to the decision boundary between clusters
    - The point could potentially belong to either cluster
    - Suggests overlapping or not well-defined clusters
* Score near -1 means:
    - The data point might be assigned to the wrong cluster
    - The point is closer to points in another cluster than its own
    - Indicates poor clustering or potential misassignments

We can use [`sklearn.metrics.silhouette_score`](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.silhouette_score.html) and [`sklearn.metrics.silhouette_samples`](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_samples.html)

````python
# silhouette scores
from sklearn.metrics import silhouette_score, silhouette_samples
````

**QUESTIONS**

* Display Silhouette score plot

````python
# ANSWER
# 1. Calcul du score moyen global (un seul chiffre)
avg_silhouette = silhouette_score(scaled_data_df, kmeans.labels_)
print(f"Average Silhouette Score: {avg_silhouette:.2f}")

# 2. Calcul du score pour chaque fleur individuelle (ncessaire pour le dessin)
sample_silhouette_values = silhouette_samples(scaled_data_df, kmeans.labels_)
````

````python
# ANSWER
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Cration de la figure
fig, ax1 = plt.subplots(1, 1)
fig.set_size_inches(10, 7)

# Limites du graphique (X = score silhouette, Y = clusters)
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, len(scaled_data_df) + (3 + 1) * 10])

y_lower = 10
# Boucle pour dessiner la "forme" de chaque cluster
for i in range(3): # k=3 clusters
    # On rcupre les scores du cluster i et on les trie
    ith_cluster_silhouette_values = sample_silhouette_values[kmeans.labels_ == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # Choix de la couleur
    color = cm.nipy_spectral(float(i) / 3)
    
    # Remplissage de la zone (le "couteau")
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Label du cluster sur la gauche
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Calcul du nouveau y_lower pour le prochain cluster
    y_lower = y_upper + 10 

# Ligne verticale rouge pour la moyenne
ax1.axvline(x=avg_silhouette, color="red", linestyle="--")

ax1.set_title("The Silhouette plot for the various clusters")
ax1.set_xlabel("The silhouette coefficient values")
ax1.set_ylabel("Cluster label")

plt.show()
````

## Task 6: Apply Hierarchical Clustering

**QUESTIONS**

* Apply Agglomerative clustering with different linkage options: complete, average, single. 

*Hint*: use [`sklearn.cluster.AgglomerativeClustering`](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html)
* For each linkage, draw a dendrogram.
* Calculate the silhouette scores

````python
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
````

````python
# --- 1. Fonction utilitaire (Standard Scikit-Learn) ---
# Cette fonction transforme le modle sklearn pour qu'il soit compatible avec dendrogram
def plot_dendrogram(model, **kwargs):
    # Cration de la matrice de liens
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # feuille (point unique)
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Trac du dendrogramme
    dendrogram(linkage_matrix, **kwargs)


# --- 2. La boucle demande (Task 6) ---
linkage_options = ['complete', 'average', 'single']

for link_method in linkage_options:
    print(f"--- Mthode : {link_method} ---")
    
    # A. Cration du modle pour le Dendrogramme (arbre complet)
    # distance_threshold=0 assure qu'on calcule tout l'arbre
    # n_clusters=None est obligatoire si distance_threshold est utilis
    model_full = AgglomerativeClustering(
        distance_threshold=0, 
        n_clusters=None, 
        linkage=link_method,
        compute_distances=True  # Important pour rcuprer les distances pour le plot
    )
    model_full.fit(scaled_data_df)
    
    # B. Affichage du Dendrogramme
    plt.figure(figsize=(10, 5))
    plt.title(f"Hierarchical Clustering Dendrogram ({link_method})")
    plot_dendrogram(model_full, truncate_mode="level", p=3)
    plt.xlabel("Number of points in node (or index of point if no parenthesis)")
    plt.show()
    
    # C. Calcul du Silhouette Score (pour k=3)
    # On recre un modle standard coup  3 clusters
    model_k3 = AgglomerativeClustering(n_clusters=3, linkage=link_method)
    labels = model_k3.fit_predict(scaled_data_df)
    
    score = silhouette_score(scaled_data_df, labels)
    print(f"Silhouette Score (k=3, {link_method}): {score:.3f}\n")
````

## Task 8: Cluster Characterisation using Apriori algorithm

Now, we would like to describe the obtained cluster. To do so, let's use frequent pattern mining and in particular **Apriori algorithm**. 

**QUESTIONS**
* First, convert numerical features to categorical (low, medium, high) based on quantiles. Add binary columns, e.g. `sepal length low`, `sepal length medium`, `sepal length high` depending on the values

````python
# ANSWER
# On cre un nouveau DataFrame vide pour stocker nos colonnes binaires
df_apriori = pd.DataFrame()

# Liste des colonnes numriques (dfinie plus tt dans votre notebook)
features = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Pour chaque caractristique numrique
for col in features:
    # 1. On dcoupe les valeurs en 3 catgories (quantiles) : low, medium, high
    # q=3 signifie "diviser en 3 parts gales"
    categories = pd.qcut(data_cleaned[col], q=3, labels=['low', 'medium', 'high'])
    
    # 2. On transforme ces catgories en colonnes binaires (One-Hot Encoding)
    # Cela va crer : 'sepal length_low', 'sepal length_medium', etc.
    dummies = pd.get_dummies(categories, prefix=col)
    
    # 3. On ajoute ces nouvelles colonnes  notre DataFrame
    df_apriori = pd.concat([df_apriori, dummies], axis=1)

# On ajoute la colonne des clusters (trouvs par K-Means) pour pouvoir trier ensuite
# Assurez-vous que la colonne 'cluster kmeans' existe bien dans data_cleaned (Task 4/5)
df_apriori['cluster'] = data_cleaned['cluster kmeans']

# Vrification visuelle
df_apriori.head()
````

To find association, we are going to use [mlxtend.frequent_patterns.apriori](https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/).

````python
# frequent patterns
from mlxtend.frequent_patterns import apriori
````

**QUESTIONS**

* Use apriori algorithm to find frequent patterns for each cluster
* Then among these itemsets, find those that are not frequent for other clusters

````python
# ANSWER
# Dictionnaire pour stocker les rsultats de chaque cluster
cluster_patterns = {}

# 1. Boucle pour trouver les patterns frquents dans chaque cluster
unique_clusters = sorted(df_apriori['cluster'].unique())

for cluster_id in unique_clusters:
    # On filtre les donnes pour ne garder que le cluster actuel
    # On supprime la colonne 'cluster' car Apriori ne veut que les attributs binaires
    subset = df_apriori[df_apriori['cluster'] == cluster_id].drop(columns=['cluster'])
    
    # On lance Apriori
    # min_support=0.5 signifie : "le motif doit apparatre dans au moins 50% des fleurs du cluster"
    frequent_itemsets = apriori(subset, min_support=0.5, use_colnames=True)
    
    # On sauvegarde le rsultat
    cluster_patterns[cluster_id] = frequent_itemsets

# 2. Comparaison pour trouver les caractristiques UNIQUES (discriminantes)
for cluster_id in unique_clusters:
    print(f"\n--- Caractristiques distinctives du Cluster {cluster_id} ---")
    
    # On rcupre tous les patterns trouvs dans les AUTRES clusters
    other_patterns = set()
    for other_id in unique_clusters:
        if other_id != cluster_id:
            other_patterns.update(cluster_patterns[other_id]['itemsets'])
    
    # On cherche les patterns du cluster actuel qui NE SONT PAS dans les autres
    current_patterns = cluster_patterns[cluster_id]['itemsets']
    
    # Filtrage
    distinctive_patterns = [p for p in current_patterns if p not in other_patterns]
    
    if distinctive_patterns:
        for p in distinctive_patterns:
            print(p)
    else:
        print("Aucun pattern unique trouv (essayez d'ajuster le min_support).")
````
